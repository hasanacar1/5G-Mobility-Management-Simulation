```python
data = pd.read_csv("C:/Users/HASAN/Desktop/hasan/itü/bitirme projesi/simülasyonlar/5G_cellular system_version_1.1/cellular_data.csv")
data.head(20)
```

```python
data = data.replace(data.loc[14]['m'], 100)
data = data.replace(data.loc[14]['b'], -100)
data.head(100)
```

```python
col_num = [] #liste

col_tar = 'margin' #string

for i in data.columns:
    if 'margin' != i :
        col_num.append(i)
    

print(col_num)


X = data[col_num]
y = data[col_tar]

X.head(40)

```

```python
X_train,X_test,y_train,y_test = ms.train_test_split(X,y,test_size=0.1)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression().fit(X_train, y_train)
```

```python
r_sq = model.score(X_train, y_train)
print('coefficient_1 of determination:', r_sq)


print('intercept:', model.intercept_)


print('slope:', model.coef_)
```

```python
y_pred = model.predict(X_test)
print('predicted response:', y_pred, sep='\n')
```
